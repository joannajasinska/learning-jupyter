```python
class Rynek:
    cenaZaM2=0
    wzrost = 0
    
    def wycenaMieszkania(self, mieszkanie):
        return self.cenaZaM2 * mieszkanie.powierzchnia
    
    def wycenMieszkaniePoLatach(self, mieszkanie, lata):
        wsp = pow((1+self.wzrost),lata)
        return self.wycenaMieszkania(mieszkanie)*wsp
    
    
class Mieszkanie:
    powierzchnia = 0
    
```

```python
rynekWarszawa = Rynek()
rynekWarszawa.cenaZaM2 = 12500
rynekWarszawa.wzrost = 0.05

mieszkanieOgrodowa = Mieszkanie()
mieszkanieOgrodowa.powierzchnia = 60

cena = rynekWarszawa.wycenaMieszkania(mieszkanieOgrodowa)
print ("cena="+str(cena))
```

```python
cenaPo5Latach = rynekWarszawa.wycenMieszkaniePoLatach(mieszkanieOgrodowa,5)
print("Cena po 5 latach: " +str(cenaPo5Latach))
```

```python
mieszkanieAnielewicza = Mieszkanie()
mieszkanieAnielewicza.powierzchnia = 56

cena2 = rynekWarszawa.wycenaMieszkania(mieszkanieAnielewicza)
cenaPo5Latach2 = rynekWarszawa.wycenMieszkaniePoLatach(mieszkanieAnielewicza,5)
```

```python
print("Ceny mieszkania na Ogrodowej przez 5 kolejnych lat:")
for x in range (1,6):
    cenaPoXLatach = rynekWarszawa.wycenMieszkaniePoLatach(mieszkanieOgrodowa,x) 
    print(cenaPoXLatach) 
```
