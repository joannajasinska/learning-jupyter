```julia
using Turing

@model function coinflip(N, successes)
    p ~ Beta(1, 1)
    successes ~ Binomial(N, p)
end

iterations = 1000
ϵ = 0.05
τ = 10

# Start sampling.
chain = sample(coinflip(100, 40), HMC(ϵ, τ), iterations)
plot(chain)
```
